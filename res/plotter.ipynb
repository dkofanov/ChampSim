```python
import json
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib as mpl
import numpy as np

TRACES = [
"600",
"602",
"603",
"605",
"607",
"619",
"620",
"621",
"623",
"625",
"627",
"628",
"631",
"638",
"641",
"644",
"648",
"649",
"654",
"657",
]


IS_CONDITIONAL_BIT = 1 << 0
TRACE_TAKEN_BIT = 1 << 1
DIR_CORRECT_BIT = 1 << 2
TGT_CORRECT_BIT = 1 << 3

COND_00 = IS_CONDITIONAL_BIT
COND_01 = IS_CONDITIONAL_BIT | TGT_CORRECT_BIT
COND_10 = IS_CONDITIONAL_BIT | DIR_CORRECT_BIT
COND_11 = IS_CONDITIONAL_BIT | DIR_CORRECT_BIT | TGT_CORRECT_BIT

RESULTS70 = {}
RESULTS90 = {}

class TraceSummary(object):
    pass

class BranchSummary(object):
    pass


def AnalyzeDetailedHistory(obj, perc_threshold=70):
    obj["cond_missed_dir"][perc_threshold] = []
    obj["cond_missed_tgt"][perc_threshold] = []


def PlotBadCondBranches(npdetails, npdetails_ips, msprd_threshold, trace_name):
    cmaplist = [(1,1,1, 1.0) for i in range(17)]
    
    # correct dir, correct tgt
    cmaplist[IS_CONDITIONAL_BIT | TRACE_TAKEN_BIT | DIR_CORRECT_BIT | TGT_CORRECT_BIT] = (0, 1, 0, 1.0)
    cmaplist[IS_CONDITIONAL_BIT | DIR_CORRECT_BIT | TGT_CORRECT_BIT] = (0, 1, 0, 1.0)

    # incorrect dir, correct tgt
    cmaplist[IS_CONDITIONAL_BIT | TRACE_TAKEN_BIT | TGT_CORRECT_BIT] = (1, 0, 0, 1.0)
    cmaplist[IS_CONDITIONAL_BIT |  TGT_CORRECT_BIT] = (1, 0, 0, 1.0)

    # correct dir, incorrect tgt
    cmaplist[IS_CONDITIONAL_BIT | TRACE_TAKEN_BIT | DIR_CORRECT_BIT] = (0, 0, 1, 1.0)
    cmaplist[IS_CONDITIONAL_BIT |  DIR_CORRECT_BIT] = (0, 0, 1, 1.0)

    # incorrect dir, incorrect tgt
    cmaplist[IS_CONDITIONAL_BIT | TRACE_TAKEN_BIT] = (1,0,1, 1.0)
    cmaplist[IS_CONDITIONAL_BIT] = (1,0,1, 1.0)
    cmaplist[16] = (0.3,0.3,0.3, .7)

    # create the new map
    cmap = mpl.colors.LinearSegmentedColormap.from_list(
        'Custom cmap', cmaplist, len(cmaplist))
    
    # define the bins and normalize
    bounds = np.linspace(0, len(cmaplist), len(cmaplist)+1)
    norm = mpl.colors.BoundaryNorm(boundaries=bounds, ncolors=len(cmaplist))

    plt.figure(figsize=(8, len(npdetails) * 0.1))
    plt.imshow(npdetails, cmap=cmap, norm=norm, interpolation='nearest', aspect='auto', extent=(0, len(npdetails[0]), 0, len(npdetails)))
    plt.xlim((0, 100))
    #plt.ylim((500, 600))
    plt.title("Hard-to-predict conditional branches statistics (miss >{}%, trace_name:\"{}\")".format(msprd_threshold, trace_name))
    plt.xlabel('Branch history')
    plt.ylabel('Found branches')
    plt.legend(handles=[
        mpatches.Patch(color=(0, 1, 0, 1.0), label='+dir, +tgt'),
        mpatches.Patch(color=(1, 0, 0, 1.0), label='-dir, +tgt'),
        mpatches.Patch(color=(0, 0, 1, 1.0), label='+dir, -tgt'),
        mpatches.Patch(color=(1, 0, 1, 1.0), label='-dir, -tgt'),
        mpatches.Patch(color=(0.3, 0.3, 0.3, .7), label='padding for alignment'),
        ])
    plt.show()


def AnalyzeDetailedHistory(obj, trace_name):
    hist = sorted(obj["details"], key=lambda lh: len(obj["details"][lh]))
    max_hlen = len(obj["details"][hist[-1]])
    
    summary = TraceSummary()
    summary.total_branches = len(hist)
    summary.msprd_branches = {}

    total_count_COND_01 = 0
    total_count_COND_10 = 0
    total_count_COND_00 = 0
    for ip in hist:
        total_count_COND_01 += np.count_nonzero(np.array(obj["details"][ip]) & COND_11 == COND_01)
        total_count_COND_10 += np.count_nonzero(np.array(obj["details"][ip]) & COND_11 == COND_10)
        total_count_COND_00 += np.count_nonzero(np.array(obj["details"][ip]) & COND_11 == COND_00)

    summary.total_count_COND_01 = total_count_COND_01
    summary.total_count_COND_10 = total_count_COND_10
    summary.total_count_COND_00 = total_count_COND_00
    
    for msprd_threshold in (70, 90):
        summary.msprd_branches[msprd_threshold] = BranchSummary()
        npdetails = []
        npdetails_ips = []
        
        for ip in hist:
            arr = np.array(sorted(obj["details"][ip]), dtype=np.int8)
            count_msprd = np.count_nonzero(arr & COND_11 == COND_00) + np.count_nonzero(arr & COND_11 == COND_01)
            if count_msprd * 100/len(arr) > msprd_threshold:
                if len(arr) < max_hlen:
                    # add padding:
                    arr = np.append(arr, np.zeros(max_hlen - len(arr), dtype=np.int8) + 16 )
                npdetails.append(arr)
                npdetails_ips.append(ip)
        
        count_COND_01 = 0
        count_COND_10 = 0
        count_COND_00 = 0
        for branch_stats in npdetails:
            count_COND_01 += np.count_nonzero(branch_stats & COND_11 == COND_01)
            count_COND_10 += np.count_nonzero(branch_stats & COND_11 == COND_10)
            count_COND_00 += np.count_nonzero(branch_stats & COND_11 == COND_00)

        summary.msprd_branches[msprd_threshold].count_COND_01 = count_COND_01
        summary.msprd_branches[msprd_threshold].count_COND_10 = count_COND_10
        summary.msprd_branches[msprd_threshold].count_COND_00 = count_COND_00

        if (len(npdetails) > 0):
            PlotBadCondBranches(npdetails, npdetails_ips, msprd_threshold, trace_name)

    print("Trace \"{}\" summary:".format(trace_name))
    cur_perc = 70
    print("  Ratio of branches with >{}% mispredict chance:".format(cur_perc))
    maj = summary.msprd_branches[cur_perc].count_COND_00
    tot = summary.total_count_COND_00
    RESULTS70[trace_name] = [0, tot]
    print("    Mispredict (-dir,-tgt) at {}%: {}/{} = {:.2f}%".format(cur_perc, maj, tot, 0 if tot == 0 else maj/tot*100))
    maj = summary.msprd_branches[cur_perc].count_COND_01
    tot = summary.total_count_COND_01
    print("    Mispredict (-dir,+tgt) at {}%: {}/{} = {:.2f}%".format(cur_perc, maj, tot, 0 if tot == 0 else maj/tot*100))
    RESULTS70[trace_name][0] += maj
    RESULTS70[trace_name][1] += tot
    maj = summary.msprd_branches[cur_perc].count_COND_10
    tot = summary.total_count_COND_10
    RESULTS70[trace_name][1] += tot
    print("    Mispredict (+dir,-tgt) at {}%: {}/{} = {:.2f}%".format(cur_perc, maj, tot, 0 if tot == 0 else maj/tot*100))

    cur_perc = 90
    print("  Ratio of branches with >{}% mispredict chance:".format(cur_perc))
    maj = summary.msprd_branches[cur_perc].count_COND_00
    tot = summary.total_count_COND_00
    RESULTS90[trace_name] = [0, tot]
    print("    Mispredict (-dir,-tgt) at {}%: {}/{} = {:.2f}%".format(cur_perc, maj, tot, 0 if tot == 0 else maj/tot*100))
    maj = summary.msprd_branches[cur_perc].count_COND_01
    tot = summary.total_count_COND_01
    RESULTS90[trace_name][0] += maj
    RESULTS90[trace_name][1] += tot
    print("    Mispredict (-dir,+tgt) at {}%: {}/{} = {:.2f}%".format(cur_perc, maj, tot, 0 if tot == 0 else maj/tot*100))
    maj = summary.msprd_branches[cur_perc].count_COND_10
    tot = summary.total_count_COND_10
    RESULTS90[trace_name][1] += tot
    print("    Mispredict (+dir,-tgt) at {}%: {}/{} = {:.2f}%".format(cur_perc, maj, tot, 0 if tot == 0 else maj/tot*100))



for trace in TRACES:    
    with open(trace + ".json") as f:
        d = json.load(f)
        obj = d[0]["roi"]["cores"][0]
        obj["trace_id"] = trace
        obj["msprd_all_types"] = 0

        for key in obj["mispredict"]:
            obj["msprd_all_types"] += obj["mispredict"][key]

        obj["cond_missed_dir"] = {}
        obj["cond_missed_tgt"] = {}

        AnalyzeDetailedHistory(obj, trace)

        #RESULTS.append(obj)
             

print("|Trace name|", end="")
for key in RESULTS70:
    print(key, end='|')
print("\n|-|", end="")
for key in RESULTS70:
    print("-", end='|')
print("\n|70%-rate|", end="")
for key in RESULTS70:
    print("{:.2g}%".format(0 if RESULTS70[key][1] == 0 else RESULTS70[key][0] * 100/RESULTS70[key][1]), end='|')
print("\n|90%-rate|", end="")
for key in RESULTS90:
    print("{:.2g}%".format(0 if RESULTS90[key][1] == 0 else RESULTS90[key][0] * 100/RESULTS90[key][1]), end='|')
print("")
```
